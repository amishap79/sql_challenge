```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import psycopg2

from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5432/sql_challenge')
connection = engine.connect
```

```python
salary_data = engine.execute("select salary  from salaries where salary > 0  order by salary")

salary_int_list = []
previous_salary = 0
count = 0
for salary in salary_data:
    salary_int = int(salary[0])
    count += 1
    if (salary_int != previous_salary):
        salary_int_list.append([previous_salary, count])
        count = 0;
        previous_salary = salary_int
        
print(salary_int_list)     

# bins = [0, 30000, 60000, 90000, 120000, 150000 ]
# count = 0
# salary_sum = 0
# salary_group = []

# for salary in salary_data:
    
#     salary_int = int(salary[0])
#     print(str(count) + "::" + str(salary))
#     if (1+count % 30000 == 0):
#         salary_sum += salary_int
#         salary_group.append(salary_sum)
#     else:
#         salary_sum += salary_int
        
#     count += 1

# print(salary_group)
    
```
